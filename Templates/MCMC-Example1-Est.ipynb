**MCMC-Example**<br>
Bimodal distribution<br>

F = $p \phi_1 + (1 - p) \phi_2$ 


```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

#%matplotlib inline
#plt.rcParams['figure.figsize'] = (17.0, 4.0)
```

```python
#F = p*phi1 + (1-p)*phi2
p = 0.5
mu1, sigma1 = 5, 2.5
mu2, sigma2 = 15, 1.5

def Fdist(x):
    return p*norm.pdf(x,mu1,sigma1) + (1-p)*norm.pdf(x,mu2,sigma2)

X = np.linspace(0, 20, 20)

fig, ax = plt.subplots(1,1)
ax.plot(X, Fdist(X), 'o-', lw=5, alpha=0.5, label='TrueDistribution')
plt.show()
```

```python
# This Function returns True if the coin with probability P of heads comes heads when flipped.
def random_coin(p):
    unif = random.uniform(0, 1)
    if unif >= p:
        return False
    else:
        return True
    
# This Function runs the MCMC chain for Beta Distribution.
def beta_mcmc(N_hops):
    states = []
    cur = random.uniform(0, 20)
    for i in range(0,N_hops):
      states.append(cur)
      next =  random.uniform(0, 20)
      ap = min(Fdist(next)/Fdist(cur), 1)
      if random_coin(ap):
            cur = next
    return states[-1000:] # Returns the last 100 states of the chain

fig, ax = plt.subplots(1,1)
plt.plot(X, Fdist(X), 'o-', lw=5, alpha=0.5, label='TrueDistribution')
plt.hist(beta_mcmc(1000), density=True, bins=25, histtype='step', label='MCMC_Distribution')
plt.show()
```

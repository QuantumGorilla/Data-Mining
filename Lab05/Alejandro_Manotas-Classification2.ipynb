```python
# Define and Assign features and label variables
color=['Red', 'Red', 'Red', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Red', 'Red']
tipo=['Sports', 'Sports', 'Sports', 'Sports', 'Sports', 'SUV', 'SUV', 'SUV', 'SUV', 'Sports']
origin=['Domestic', 'Domestic', 'Domestic', 'Domestic', 'Imported', 'Imported', 'Imported', 'Domestic', 'Imported', 'Imported']
stolen=['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No','Yes']
```

```python
# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting color string labels into numbers.
color_encoded=le.fit_transform(color)
```

```python
# Converting tipo string labels into numbers
tipo_encoded=le.fit_transform(tipo)

# Converting origin string labels into numbers
origin_encoded=le.fit_transform(origin)

# Converting stolen string labels into numbers
stolen_encoded=le.fit_transform(stolen)

#Print color_encoded
print("Color:", color_encoded)

# Print tipo encoded
print("Tipo:", tipo_encoded)

# Print origin encoded
print("origin:", origin_encoded)

#Print stolen
print("Stolen:", stolen_encoded)
```

```python
#Combinig color and tipo into single list of tuples and assign the features
features = list(zip(color_encoded,tipo_encoded,origin_encoded))

# Print features
print(features)
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier and assign to model
model = GaussianNB()

# Train the model using the training sets
model.fit(features,stolen_encoded)

#Call Predict function for 
predicted = model.predict([[0,0,0]])

#Print predicted value 0: Red, 0: SUV and 0: Domestic
print("Predicted value:", predicted)
```

Es 0 lo que indica que un carro SUV domestico tiene 0 probabilidades de ser robado.
